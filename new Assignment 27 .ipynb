```python
#1.Extract only the alphabetic characters from a string containing letters, digits, and symbols.
s = input("Enter a string: ")
result = ""
for ch in s:
    if ch.isalpha():
        result += ch
print(result)
```

```python
#2.Normalize a string so multiple spaces between words become a single space.
s = input("Enter a string: ")
words = s.split()  
result = " ".join(words)
print(result)
```

```python
#3.Check if two strings are anagrams while ignoring case and spaces.
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
if sorted(s1) == sorted(s2):
    print("Anagram")
else:
    print("Not an Anagram")
```

```python
#4.Reverse a string without using slicing.
string = input("Enter a string: ")
reverse = ""
for char in string:
    reverse = char + reverse
print("Reversed string:", reverse)
```

```python
#5.Count the number of sentences in a paragraph ending with '.', '?', or '!'.
paragraph = input("Enter paragraph: ")
count = 0
for char in paragraph:
    if char in ".?!":
        count += 1
print(count)
```

```python
#6.Find the second most frequent character in a string.
s = input("Enter a string: ")
freq = {}
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1
second = sorted(freq, key=freq.get, reverse=True)[1]
print(second)
```

```python
#7. Compress a string like 'aaabbccccdde' into 'a3b2c4d2e1'.
s = "aaabbccccdde"
result = ""
count = 1
for i in range(len(s)-1):
    if s[i] == s[i+1]:
        count += 1
    else:
        result += s[i] + str(count)
        count = 1
result += s[-1] + str(count)
print(result)
```

```python
#8. Check if a string contains a repeating substring pattern.
s = input("Enter string: ")
if s and s in (s+s)[1:-1]:
    print("Repeating substring pattern")
else:
    print("No repeating substring pattern")
```

```python
#9.Remove characters that appear more than once in a string.
string = input("Enter a string: ")
result = ""
for char in string:
    if string.count(char) == 1:
        result += char
print(result)
```

```python
#10.. Split a camelCase string into separate words.
string = input("Enter camelCase string: ")
result = ""
for char in string:
    if char.isupper():
        result += " " + char
    else:
        result += char
print(result.strip())
```

```python
#11.Rotate a string to the right by k positions using slicing only once.
s = input("Enter string: ")
k = int(input("Enter k: "))
k = k % len(s)
rotated = s[-k:] + s[:-k]
print("Rotated string:", rotated)
```

```python
#16. Convert a string into alternating case starting with uppercase.
s = input("Enter a string:")
r = ""
u = True
for c in s:
    if c.isalpha():
        r += c.upper() if u else c.lower()
        u = not u
    else:
        r += c
print(r)
```

```python
#17. Count words in a sentence that start and end with the same letter.
s = input("Enter a sentence:")
count = 0
for w in s.split():
    if w[0].lower() == w[-1].lower():
        count += 1
print(count)
```

```python
#18. Swap every pair of characters in a string.
s = input("Enter a string: ")
result = ""
for i in range(0, len(s), 2):
    if i + 1 < len(s):
        result += s[i+1] + s[i]
    else:
        result += s[i]          
print("Swapped string:", result)
```

```python
#19. Clean comma-separated emails: trim, lowercase, unique, sorted.
s = input("Enter emails: ")
emails = s.split(",")
clean = sorted(set(e.strip().lower() for e in emails))
print(",".join(clean))
```

```python
#23.Implement a custom split function without using split().
text = input("Enter text: ")
sep = input("Enter separator: ")
word = ""
words = []
for ch in text:
    if ch == sep:
        words = words + [word]   
        word = ""
    else:
        word += ch
words = words + [word]       
print(words)
```

```python
#24.Rearrange a string so lowercase letters appear first, then uppercase.
s = input("Enter string: ")
lower = ""
upper = ""
for ch in s:
    if ch.islower():
        lower += ch
    elif ch.isupper():
        upper += ch
print(lower + upper)
```

```python
#25.Generate all unique subsets of characters in lexicographical order.
s = input("Enter string: ")
s = sorted(s)
subs = [""]
for ch in s:
    new = []
    for sub in subs:
        new += [sub + ch]
    subs += new
subs = sorted(set(subs))
for sub in subs[1:]:   
    print(sub)
```

```python
#29.Remove all vowels from a string while preserving character order.
s = input("Enter string: ")
v = "aeiouAEIOU"
result = ""
for ch in s:
    if ch not in v:
        result += ch
print("Result:", result)
```
