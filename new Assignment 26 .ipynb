```python
#1.
n=input("enter a password:-")
upper=False
Lower=False
Digit=False
count=0
for i in n:
    count +=1
    if i.isupper():
        upper=True
    if i.islower:
        Lower=True
    if i.isdigit:
        Digit=True
if count>=8 and upper and Lower and Digit:
    print("Password is valid")
else:
    print("Password is invalid")
```

```python
#2.
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")
if sorted(str1) == sorted(str2):
    print("The strings are anagrams.")
else:
    print("The strings are not anagrams.")
```

```python
#3.
s = input("Enter a 3-letter string: ")
if len(s) == 3:
    if s[0] == s[1] == s[2]:
        print("Pattern: AAA")
    elif s[0] == s[2]:
        print("Pattern: ABA")
    else:
        print("Pattern: ABC")
else:
    print("Please enter exactly 3 letters.")
```
