```python
#label encoding
from sklearn.preprocessing import LabelEncoder
flowers = ["rose" "lilly","sunflower","daisy","tuli","orchid","carnation"]
encoder = LabelEncoder()
flowers_encoded = encoder.fit_transform(flowers)
print(flowers_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
#one hot encoding using Pandas 
import pandas as pd 
data={
    "student":["aa","bb","cc","dd"],
    "grade":["A","A+","B","A"],
    "course":["bca","bba","bsc","bca"],
    "city":["pune","satara","mumbai","pune"],
    "marks":[90,91,92,93]

}
df=pd.DataFrame(data)
print(df)
encoded_df=pd.get_dummies(df,columns=["student","grade","course","city","marks"],dtype=int)
print(encoded_df)
    
```

```python
#ordinal Encoding 
from sklearn.preprocessing import OrdinalEncoder
size = [["m"], ["s"], ["l"], ["m"], ["s"], ["xl"], ["xxl"]]
encoder = OrdinalEncoder(categories=[["s", "m", "l", "xl", "xxl"]])
encoded_size = encoder.fit_transform(size)
print(encoded_size)
```
